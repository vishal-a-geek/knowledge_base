#### Chained Indexing

```python
weather_df[0:10:2][['Date/Time (formatted)', 'Visibility (km)']]
```

```python
weather_df[['Date/Time (formatted)', 'Visibility (km)']][0:10:2]
```

When using chained indexing, the order and type of indexing operation partially determine whether the result is 
- a slice into original object
- or a copy of slice

```python
# first indexing
df1 = weather_df[:6:2]

# second indexing
df2 = df1[['Rel Hum (%)', 'Visibility (km)']]
```

While both results are correct in this **read-only** case, chained indexing may give unpredicatble behaviours when writing to a dataframe. This is because indexing could either return 
- a "view" (of slices of the dataframe)
- or a copy of dataframe

Thus, while working with a large dataset, avoid slicing and rather use `loc` or `iloc`

### `loc` and `iloc`

- `.loc`: indexer based on label-location for selection by label
- `.iloc`: indexer based on integer-location for selection by position

```python
weather_df.loc[0:5, ['Visibility (km)', 'Rel Hum (%)']] ## label based indexing (0 to 5)
```

```python
weather_df[['Visibility (km)', 'Rel Hum (%)']].iloc[0:5] ## position based indexing (0 to 4)
```

## Summary / Cheatsheet: Selection/Indexing/Filtering
This is a handy reminder for what syntax will get what result.

Syntax | Function | Remarks
:--- | :--- | :---
**`df['some_label']`** |  Get the (single) Column referenced by name `some_label` | A **str** is provided
**`df[['label1', 'label2']]`** | Get multiple columns referenced by given names | A **list** is provided 
**`df[start:end:step]`** | Get corresponding rows (same as list slicing) | A **slicing operator**<br/> is provided
**`df[boolean array/df]`** | Get corresponding rows (same as list slicing) | A **filter object** is provided
**`df.loc [row_sel, col_sel]`** | Select specified rows and columns (by labels) | 
**`df.iloc[row_sel, col_sel]`** | Select specified rows and columns (by index) |

## Concat, Merging, Join DF

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
}, index=[0, 1, 2, 3])

df2 = pd.DataFrame({
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']
}, index=[4, 5, 6, 7])

df3 = pd.DataFrame({
    'A': ['A8', 'A9', 'A10', 'A11'],
    'B': ['B8', 'B9', 'B10', 'B11'],
    'C': ['C8', 'C9', 'C10', 'C11'],
    'E': ['D8', 'D9', 'D10', 'D11']
}, index=[8, 9, 10, 11])
```

```python
# Join
left_df = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
}, index=['K0', 'K1', 'K2']) 

right_df = pd.DataFrame({
    'C': ['C0', 'C2', 'C3'],
    'D': ['D0', 'D2', 'D3']
}, index=['K0', 'K2', 'K3'])
```

```python
left_df
```

```python
left_df.join(right_df, how="outer")
```

```python
left_df.join(right_df, how="inner")
```

```python
# Merging on multiple keys
left = pd.DataFrame({
    'key1': ['K0', 'K0', 'K1', 'K2'],
    'key2': ['K0', 'K1', 'K0', 'K1'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})
    
right = pd.DataFrame({
    'key1': ['K0', 'K1', 'K1', 'K2'],
    'key2': ['K0', 'K0', 'K0', 'K0'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})
```

```python
df = pd.DataFrame({
    "name": ["Tom", "Andy", "Lucas"],
    "DoB": ["08-05-1997", "04-28-1996", "12-16-1995"],
})
dw_mapping = {
    0: "Monday",
    1: "Tuesday",
    2: "Wednesday",
    3: "Thursday",
    4: "Friday",
    5: "Saturday",
    6: "Sunday",
}
today = pd.to_datetime('today')

df["DoB (formatted)"] = pd.to_datetime(df["DoB"])
df["Year"] = df["DoB (formatted)"].dt.year
df["Month"] = df["DoB (formatted)"].dt.month
df["Day"] = df["DoB (formatted)"].dt.day
df["week_of_year"] = df["DoB (formatted)"].dt.weekday
df["day_of_week"] = df["DoB (formatted)"].dt.day_of_week
df["is_leap_year"] = df["DoB (formatted)"].dt.is_leap_year
df["day_of_week_name"] = df["DoB (formatted)"].dt.weekday.map(dw_mapping)
df["age"] = pd.to_datetime('today').year - df['DoB (formatted)'].apply(lambda x:x.year-1 if (x.month, x.day)>(today.month, today.day) else x)
df
```
